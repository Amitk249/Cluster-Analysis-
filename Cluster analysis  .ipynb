```python
numeric = df.select_dtypes(include=['number'])
```

```python
from sklearn.cluster import KMeans 
```

```python
cluster = KMeans(3)
cluster.fit(numeric)
```

```python
numeric['clusterid'] = cluster.labels_
```

```python
numeric[0:5]
```

```python
markers = ['+', '^', '.']

sn.lmplot(data = numeric , x = 'Sr.NO', y = 'Price', hue = 'clusterid',fit_reg = False , markers = markers , aspect = 2 , height = 5)

```

# KMEANS clustering Using Normalization method

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
scaled_data  = scalar.fit_transform(numeric[['Sr.NO','Price']])
```

```python
scaled_data[0:5]
```

```python
from sklearn.cluster import KMeans
```

```python
clusters_New = KMeans(3, random_state = 45)
clusters_New.fit(scaled_data)
numeric['clusterid_new'] = clusters_New.labels_
```

```python
markers = ['*','^','.']
sn.lmplot(data = numeric , x = 'Sr.NO', y = 'Price', hue = 'clusterid_new',fit_reg = False , markers = markers , aspect = 2 , height = 5)
```
